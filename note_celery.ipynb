*Note about celery:*
- pip install "celery[librabbitmq,redis,auth,msgpack]"
- pip install flower
- pip install eventlet. celery is not support window, we have to use eventlet.
- celery -A celery_app worker -c 4 -l INFO -P eventlet. to start celery worker
- celery -A celery_app flower --address=localhost --port=5566. to start flower
- selenium can not start chrome driver in local machine because hosts file config.
    is the mapping between localhost name string with the actuall ip address of the local machine 127.0.0.1
